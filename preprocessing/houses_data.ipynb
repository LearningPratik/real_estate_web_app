```python
import numpy as np
import pandas as pd

import copy
```

```python
df = pd.read_csv(r"D:\Projects\plot_recommendition\data\pune\Flats\houses.csv")
df.sample(5)
```

```python
# shape
df.shape
```

```python
# info
df.info()
```

```python
# check for duplicates
df.duplicated().sum()
```

```python
df_copy = copy.deepcopy(df)
df_copy.sample(2)
```

```python
df_copy = df_copy.drop_duplicates()
```

```python
df_copy.shape
```

```python
# check for missing values
df_copy.isnull().sum()
```

```python
# Columns to drop -> property_name, link, property_id
df_copy.drop(columns=['link','property_id'], inplace=True)
```

```python
df_copy.head()
```

```python
# rename columns
df_copy.rename(columns={'rate':'price_per_sqft'},inplace=True)
df_copy.sample()
```

```python
# society
df_copy['society'].value_counts()
```

```python
df_copy['society'].value_counts().shape
```

```python
import re
df_copy['society'] = df_copy['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df_copy['society'].value_counts().shape
```

```python
df_copy['society'] = df_copy['society'].str.replace('nan','independent')
```

```python
# price
df_copy['price'].value_counts()
```

```python
df_copy = df_copy[df_copy['price'] != 'Price on Request']
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df_copy['price'] = df_copy['price'].str.split(' ').apply(treat_price)
```

```python
df_copy.head()
```

```python
# price_per_sqft
df_copy['price_per_sqft'].value_counts()
```

```python
df_copy['price_per_sqft'] = df_copy['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
df_copy = df_copy[~df_copy['bedRoom'].isnull()]
```

```python
df_copy['bedRoom'] = df_copy['bedRoom'].str.split(' ').str.get(0).astype('int')
```

```python
df_copy['bathroom'] = df_copy['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
df_copy['balcony'] = df_copy['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df_copy['additionalRoom'].fillna('not available',inplace=True)
```

```python
df_copy['additionalRoom'] = df_copy['additionalRoom'].str.lower()
```

```python
# floors
df_copy['noOfFloor'].value_counts()
```

```python
df_copy['noOfFloor'].isnull().sum()
```

```python
df_copy['noOfFloor'] = df_copy['noOfFloor'].str.split(' ').str.get(0)
```

```python
df_copy['facing'].fillna('NA',inplace=True)
```

```python
df_copy['area'] = round((df_copy['price']*10000000)/df_copy['price_per_sqft'])
```

```python
df_copy.insert(loc=1,column='property_type',value='house')
```

```python
df_copy.sample(2)
```

```python
df_copy.to_csv('house_cleaned.csv',index=False)
```
